### 훈련 데이터와 테스트 데이터 직접 분리하기

```python
from sklearn.datasets import load_wine

data = load_wine() # 와인 데이터 불러오기

# 데이터 확인
print(data.data.shape)
print(data.target.shape)
```

```python
X_train = data.data[:142] # 142개의 훈련 데이터
X_test = data.data[142:] # 13개의 테스트 데이터 
print(X_train.shape, X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X_train = data.data[:142]
X_test = data.data[142:]
print("학습 데이터 분리 :", X_train.shape, X_test.shape)

y_train = data.target[:142]
y_test = data.target[142:]
print("라벨 데이터 분리", y_train.shape, y_test.shape)

model = RandomForestClassifier() # 모델 불러오기
model.fit(X_train, y_train) # 학습하기

y_pred = model.predict(X_test)
print("정답률=", accuracy_score(y_test, y_pred))
```

```python
# train_test_split() 사용해서 분리하기
from sklearn.model_selection import train_test_split

result = train_test_split(data.data, data.target, test_size=0.2, random_state=42)
result
```
