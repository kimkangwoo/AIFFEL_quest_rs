#  2019 2nd ML month with KaKR
- 캐글 코리아와 함께하는 2nd ML 대회 - House Price Prediction
- 이번 대회의 평가 방식은 Root Mean Squared Error 입니다.

### Data fields
1. **ID**: 집을 구분하는 번호
2. **date**: 집을 구매한 날짜
3. **price**: 집의 가격 (Target variable)
4. **bedrooms**: 침실의 수
5. **bathrooms**: 화장실의 수
6. **sqft_living**: 주거 공간의 평방 피트 (면적)
7. **sqft_lot**: 부지의 평방 피트 (면적)
8. **floors**: 집의 층 수
9. **waterfront**: 집의 전방에 강이 흐르는지 유무 (a.k.a. 리버뷰)
10. **view**: 집이 얼마나 좋아 보이는지의 정도
11. **condition**: 집의 전반적인 상태
12. **grade**: King County grading 시스템 기준으로 매긴 집의 등급
13. **sqft_above**: 지하실을 제외한 평방 피트 (면적)
14. **sqft_basement**: 지하실의 평방 피트 (면적)
15. **yr_built**: 지어진 년도
16. **yr_renovated**: 집을 재건축한 년도
17. **zipcode**: 우편번호
18. **lat**: 위도
19. **long**: 경도
20. **sqft_living15**: 2015년 기준 주거 공간의 평방 피트 (면적, 집을 재건축했다면, 변화가 있을 수 있음)
21. **sqft_lot15**: 2015년 기준 부지의 평방 피트 (면적, 집을 재건축했다면, 변화가 있을 수 있음)

```python
# 데이터 불러오기 
import pandas as pd

folder_path = "./data/"
train = pd.read_csv(folder_path+'train.csv')
test = pd.read_csv(folder_path+'test.csv')

# 데이터 확인
print('훈련용 데이터 셋 : ', train.shape)
print('테스트용 데이터 셋 : ', test.shape)
```

```python
# 틀린값 찾기
set(train.keys()) -set(test.keys()) 
```

```python
y = train['price'] # 정답값 추출
train = train.drop(['price'], axis=1)
train.shape
```

```python
# 데이터 확인 
train.info() # date만 string
```

```python
# 형변환
train['date'] = train['date'].apply(lambda i: i[:6]).astype(int)
test['date'] = test['date'].apply(lambda i: i[:6]).astype(int)
train.info()
```

```python
# 필요없는 값 제거 
import matplotlib.pyplot as plt
import seaborn as sns

features = list(train.keys()) # Keys 저장

for feature in features:
    sns.kdeplot(train[feature])
    plt.title(feature)
    plt.show()
```

```python
plt.hist(y, bins=50)
plt.title('House Pricing')
plt.show()
```

```python
print('집가격 범위 :', y.min(),'~', y.max())
```

```python
corr = train
corr['price'] = y
corr = corr.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.1f', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 상관관계가 0.3이상인 것들
fea= [i[1] for i in zip(corr['price'], list(corr['price'].keys())) if i[0]>0.3 or i[0]<-0.3]
fea
```

```python
# 상관관계가 높은 것들만 data로 삼는다. 
fea.remove('price') # 정답값은 제외한다.
data = train[fea]
test_df = test[fea]
data
```

```python
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMRegressor

param_grid = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [10, 20, 30],
}
model = LGBMRegressor(random_state=42)
grid_model = GridSearchCV(model, param_grid=param_grid, \
                        scoring='neg_mean_squared_error', \
                        cv=5, verbose=1, n_jobs=16)

grid_model.fit(data, y)
grid_model.cv_results_
```

```python
params = grid_model.cv_results_['params']
score = grid_model.cv_results_['mean_test_score']
rank = grid_model.cv_results_['rank_test_score']

results = pd.DataFrame(params)
results['score'] = score
results['rank'] = rank

print(grid_model.best_params_)
results.sort_values(by="rank")
```

```python
print("최적의 하이퍼파라미터:", grid_model.best_params_)

best_model = LGBMRegressor(**grid_model.best_params_)
best_model.fit(data, y)

y_pred = best_model.predict(test_df)
print("테스트 데이터 평가 결과:", y_pred)
```
